```python
import pandas as pd 
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('Zomato_data.csv')
```

```python
df.drop('index',axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df1=df[df['num of ratings']>120]
```

```python
df1
```

```python
bar_plot_all=df['restaurant type'].value_counts().head(10)
```

```python
bar_plot_known=df1['restaurant type'].value_counts().head(10)
```

```python
plt.figure(figsize=(9,6))
plt.bar(bar_plot_all.index,bar_plot_all,color='Midnightblue')
plt.title('Restaurant count of Each type')
plt.xlabel('Restaurant type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
#the number of restaurants zomato has tied up with
```

```python
plt.figure(figsize=(9,6))
plt.bar(bar_plot_known.index,bar_plot_known,color='Midnightblue')
plt.title('Restaurant count of Each type')
plt.xlabel('Restaurant type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
#the below graph shows the  restaurants with frequently visited or ordered customer that is restaurant 
#count with restaurants having number of rating greater than 120
```

```python
tf=df[df['restaurant type']=='Quick Bites']
```

```python
tf['avg cost (two people)'].mean()
```

```python
y_cost=np.zeros(10)
for i in range(len(bar_plot_all)):
    df_rtype=df[df['restaurant type']==bar_plot_all.index[i]]
    y_cost[i]=df_rtype['avg cost (two people)'].mean()
```

```python
y_cost
```

```python
plt.figure(figsize=(9,6))
plt.bar(bar_plot_all.index,y_cost,color='Midnightblue')
plt.title('Restaurants vs Cost')
plt.xlabel('Restaurant type')
plt.ylabel('Average Cost')
plt.xticks(rotation=45)
plt.show()
#this shows the average cost of each restaurant
```

```python
sns.despine()
sns.regplot(x=df['Rating'],y=df['avg cost (two people)'],data=df,scatter_kws={'color':'k'},line_kws={'color':'red'})
plt.show()
#as rating increases cost increases but the slope is so low which states that the prices from low to high
#Cost Increases only by small amount
```

```python
y_rating=np.zeros(10)
for i in range(len(bar_plot_all)):
    df_rtype=df1[df1['restaurant type']==bar_plot_all.index[i]]
    y_rating[i]=df_rtype['Rating'].mean()
sns.set_style('white')
fig,ax=plt.subplots(figsize=(9,))
ax.bar(bar_plot_all.index,y_cost,color='black')
ax.tick_params(axis='x',rotation=45)
ax1=ax.twinx()
ax1.plot(bar_plot_all.index,y_rating,color='crimson',marker='D')
ax1.set_ylim(0,5)
ax.set_xlabel('Restaurant type',weight='bold')
ax.set_ylabel('Average Cost',weight='bold')
ax1.set_ylabel('Mean Rating',weight='bold')
plt.title('Cost Vs Mean Rating',fontsize='14',weight='bold')
plt.plot()
```
